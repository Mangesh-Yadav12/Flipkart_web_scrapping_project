```python
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlopen as urReg
```

```python
flipkart_url = "https://www.flipkart.com/search?q="+"iphone12"
```

```python
flipkart_url
```

```python
response_website = urReg(flipkart_url)
```

```python
data_flipkart = response_website.read()
```

```python
beautified_html = bs(data_flipkart,"html.parser")
```

```python
beautified_html
```

```python
bigbox = beautified_html.find_all("div",{"class":"_1AtVbE col-12-12"})
```

```python
bigbox
```

```python
product3 = "https://www.flipkart.com"+ bigbox[2].div.div.div.a["href"]
```

```python
import requests
```

```python
product33 = requests.get(product3)
```

```python
product3
```

```python
product33.encoding= "utf-8"
```

```python
product3_page = bs(product33.text, "html.parser")
```

```python
all_reviews = product3_page.find_all("div",{"class":"_16PBlm"})
```

```python
len(all_reviews)
```

```python
for i in all_reviews:
    price = product3_page.find_all("div",{"class":"_30jeq3 _16Jk6d"})[0].text
    date = i.div.div.find_all("p",{"class":"_2sc7ZR"})[1].text
    reviews = i.div.div.find_all("div",{"class":""})[0].text
    comment_header = i.div.div.div.p.text
    rating = i.div.div.div.div.text
    name = i.div.div.find_all("p",{"class":"_2sc7ZR _2V5EHH"})[0].text
    print(name," ",rating," ",date ," ",price," ",reviews," ",comment_header)
    print()
```
